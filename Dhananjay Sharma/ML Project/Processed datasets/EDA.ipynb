```python
def clean_dataset(data):
    # 1. Drop common "trash" index columns
    data = data.drop(columns=['index', 'level_0'], errors='ignore')
    
    # 2. Drop any column that starts with "Unnamed" (common in CSV exports)
    data = data.loc[:, ~data.columns.str.contains('^Unnamed', case=False)]
    
    # 3. Ensure Year/State/District stay as regular columns, not an Index
    data = data.reset_index(drop=True)
    
    return data
```

```python
# Clean all dataframes in a loop
df1 = clean_dataset(df1)
df2 = clean_dataset(df2)
df3 = clean_dataset(df3)
df4 = clean_dataset(df4)
df5 = clean_dataset(df5)
```

```python
from functools import reduce

# List of your cleaned dataframes
dfs = [df1, df2, df3, df4, df5]

# Merge them all on the common columns
df = reduce(lambda left, right: pd.merge(left, right, on=['Year', 'State Name', 'Dist Name'], how='inner'), dfs)

# Final check
df.info()
```

```python
df=pd.read_csv("final_dataset.csv")
```

```python
# Remove any column that isn't actual data
cols_to_drop = [c for c in df.columns if 'index' in c.lower() or 'unnamed' in c.lower()]
df.drop(columns=cols_to_drop, inplace=True)
```

```python
# 1. Define the columns that must stay at the start
fixed_cols = ['Year', 'State Name', 'Dist Name']

# 2. Get all other columns and sort them alphabetically
other_cols = sorted([c for c in df.columns if c not in fixed_cols])

# 3. Reorder the DataFrame
df = df[fixed_cols + other_cols]
```
